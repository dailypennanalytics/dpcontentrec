```python
tags = pd.read_csv('tags_34.csv')
```

```python
data = pd.read_csv('street.csv')
```

```python
tags = tags.rename(columns={"title_url": "slug"})
```

```python
join = pd.merge(data, tags, how='left', on = 'slug')
```

```python
join.shape
```

```python
join.head()
```

```python
join = join.drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis=1)
```

```python
join['tags'] = join['tags'].str.replace("'", "")
```

```python
join['tags'] = join['tags'].str.lower()
```

```python
join['tags'] = join['tags'].str.strip('[]').str.split(',')
```

```python
join.head()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
```

```python
#replace Na with empty list
for row in join.loc[join.tags.isnull(), 'tags'].index:
    join.at[row, 'tags'] = []
```

```python
s = join['tags']
```

```python
encoded = pd.DataFrame(mlb.fit_transform(s),columns=mlb.classes_, index=join.index)
```

```python
encoded.head()
encoded.shape
```

```python
encoded.sum(axis=0)
```

```python
columns = list(encoded)
d = []
for i in columns:
    s = encoded[i].sum()
    if (s <= 1):
        d.append(i)
```

```python
len(d)
```

```python
#drop columns
encoded = encoded.drop(d, axis=1)
```
